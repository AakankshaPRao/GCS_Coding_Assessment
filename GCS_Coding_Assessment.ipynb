```python
import requests, json

class OpenWeatherApi:
  def __init__(self):
    self.api_call = 'http://api.openweathermap.org/data/2.5/weather'
    self.api_key = 'aaadca7deffdd1a712e2bae603b42cb1'
    self.favourites = []

  def search_city_weather_details(self, name):
    response = requests.get(f"{self.api_call}?q={name}&appid={self.api_key}")

    if response.status_code == 200:
      data = response.json()
      self.display_city_current_weather(data)
    else:
      print(f"Error: Cannot find weather details for {name.capitalize()}")

  def display_city_current_weather(self, data):
    name = data['name']
    weather = data['weather'][0]['description']
    temperature = data['main']['temp']
    humidity = data['main']['humidity']
    print(f"Weather in {name.capitalize()} is {weather} with temperature {temperature}°C and humidity is {humidity}%")

  def add_city_to_favourites(self, name):
    if len(self.favourites) < 3:
      if name not in self.favourites:
        self.favourites.append(name.lower())
        print(f"{name.capitalize()} added to your favourites list.")
      else:
        print(f"{name.capitalize()} is already in your favourites list.")
    elif len(self.favourites) == 3:
      if name in self.favourites:
        print(f"{name.capitalize()} is already in your favourites list. Also, favourites list is full, remove a city before adding a new one.")
      else:
        print("Favourites list is full. Remove a city before adding a new one.")

  def list_favourite_cities(self):
    print("Favourites List:")
    for name in self.favourites:
      self.search_city_weather_details(name)

  def update_favourite_cities(self, remove_city, add_city):
    if remove_city in self.favourites:
      self.favourites.remove(remove_city)
      print(f"{remove_city.capitalize()} removed from your favourites list.")

    else:
      print(f"{remove_city} is not in your favourites list.")

    add_new_city = input(f"Do you want to add {add_city.capitalize()} in your favourites list? (yes/no): ").lower()
    if add_new_city == 'yes':
      self.add_city_to_favourites(add_city)
    elif add_new_city == 'no':
      print("Favourites list remains unchanged.")
    else:
      print("Invalid input.")


if __name__ == "__main__":
  open_weather_api = OpenWeatherApi()
  print("\n********* Welcome to Open Weather Api Application *********")

  while True:
    print("\n1. Search for weather details of a city\n2. Add a new city to your favourites list\n3. Display your top 3 favourite cities\n4. Update your favourite cities list\n5. Exit\n")

    choice = input("Enter your choice from 1 to 5 : ")

    if choice == '1':
      name = input("Enter the name of the city to search: ")
      open_weather_api.search_city_weather_details(name.lower())
    elif choice == '2':
      name = input("Enter the name of the city to add to your favourites list: ")
      open_weather_api.add_city_to_favourites(name.lower())
    elif choice == '3':
      if len(open_weather_api.favourites) == 0:
        print("Your favourites list is empty! Add a city to your list to display your favourites list.")
      else:
        open_weather_api.list_favourite_cities()
    elif choice == '4':
      if len(open_weather_api.favourites) == 0:
        print("Your favourites list is empty! Add a city to your list.")
      else:
        remove_city = input("Enter the name of the city to remove from favourites: ")
        add_city = input("Enter the name of the city to add to favourites: ")
        open_weather_api.update_favourite_cities(remove_city.lower(), add_city.lower())
    elif choice == '5':
      print("You have closed the city weather application!")
      break
    else:
      print("Invalid choice.")
```

```python
def sort_and_find_median(numbers):
    sort(numbers)

    n = len(numbers)
    if n % 2 == 0:
        return (numbers[n//2 - 1] + numbers[n//2]) / 2
    else:
        return numbers[n//2]

def sort(numbers):
    if len(numbers) > 1:
        mid = len(numbers) // 2
        left = numbers[:mid]
        right = numbers[mid:]
        i = j = k = 0

        sort(left)
        sort(right)

        while i < len(left) and j < len(left):
            if left[i] < right[j]:
                numbers[k] = left[i]
                i += 1
            else:
                numbers[k] = right[j]
                j += 1
            k += 1

        while i < len(left):
            numbers[k] = left[i]
            i += 1
            k += 1

        while j < len(right):
            numbers[k] = right[j]
            j += 1
            k += 1

input = [6, 8, 0, 4, 1]
median = sort_and_find_median(input)
print("Sorted Array:", input)
print("Median:", median)
```
